```python
def add(x,y):
  return x+y
def substract(x,y):
  return x-y
def multiply(x,y):
  return x*y
def divide(x,y):
  if y==0:
    return "division by zero is not allowed"
  else:
    return x/y
def get_input():
  try:
    x=float(input("enter the first number"))
    y=float(input("enter the second number"))
    return x,y
  except ValueError:
    print("invalid input!please enter valid numbers.")
    return get_input() # Recursive call to get valid input
def print_menu():
  print("\nselect operation:")
  print("1. add")
  print("2. substract")
  print("3. multiply")
  print("4. divide")
  print("5. exit")
def calculator():
   while True:
    print_menu()
    choice=input("enter your choice(1/2/3/4/5):")
    if choice=="5":
      print("exiting calculator")
      break

    if choice in ("1", "2", "3", "4"):
      x,y=get_input()
      if choice=="1":
        print("result:",add(x,y))
      elif choice=="2":
        print("result:",substract(x,y))
      elif choice=="3":
        print("result:",multiply(x,y))
      elif choice=="4":
        print("result:",divide(x,y))
    else:
      print("invalid choice!please select a valid option (1/2/3/4/5)")

calculator()
print("END")
```

```python
def add(x,y):
  return x+y
def substract(x,y):
  return x-y
def multiply(x,y):
  return x*y
def divide(x,y):
  if y==0:
    return "division by zero is not allowed"
  else:
    return x/y
def get_input():
  try:
    x=float(input("enter the first number"))
    y=float(input("enter the second number"))
    return x,y
  except ValueError:
    print("invalid input!please enter valid numbers.")
    return get_input() # Recursive call to get valid input
def print_menu():
  print("\nselect operation:")
  print("1. add")
  print("2. substract")
  print("3. multiply")
  print("4. divide")
  print("5. exit")
def calculator():
   while True:
    print_menu()
    choice=input("enter your choice(1/2/3/4/5):")
    if choice=="5":
      print("exiting calculator")
      break
      x,y=get_input()
      if choice=="1":
        print(f"{x}+{y}={add(x,y)}")
      elif choice=="2":
        print(f"{x}-{y}={substract(x,y)}")
      elif choice=="3":
        print(f"{x}*{y}={multiply(x,y)}")
      elif choice=="4":
        print(f"{x}/{y}={divide(x,y)}")
    else:
      print("invalid choice!please select a valid option (1/2/3/4/5)")

calculator()
print("END")
```
