```python
import numpy as np

#Rainfall data over 12 months

n=int(input("Enter the number of months:")) #12
rainfall_data=[]

for month in range(n):
  data=float(input("Enter rainfall(mm):"))
  rainfall_data.append(data)

rainfall=np.array(rainfall_data)
print(rainfall)

#Corresponding yield data over 12 months

yield_info=[]

for month in range(n):
  yields=float(input("Enter yield(tons/ha):"))
  yield_info.append(yields)

yield_data=np.array(yield_info)
print(yield_data)
```

```python
import numpy as np

print('===RAINFALL STATISTICS===')
print(f'Mean:{np.mean(rainfall):.2f}mm')
print(f'Median:{np.median(rainfall):.2f}mm')
print(f'Std Dev:{np.std(rainfall):.2f}mm')
print(f'Variance:{np.var(rainfall):.2f}')
print(f'Min:{np.min(rainfall)}mm(Month{np.argmin(rainfall)+1})')
print(f'Max:{np.max(rainfall)}mm(Month{np.argmax(rainfall)+1})')
print(f'Range:{np.ptp(rainfall)}mm')

print('\n===YIELD STATISTICS===')
print(f'Mean:{np.mean(yields):.2f}tons/ha')
print(f'Median:{np.median(yields):.2f}tons/ha')
print(f'Std Dev:{np.std(yields):.2f}tons/ha')

print('\n===PERCENTILES===')
percentiles=[25,50,75,90]
rainfall_percentiles=np.percentile(rainfall,percentiles)
for p,val in zip(percentiles,rainfall_percentiles):
  print(f'{p}th percentile rainfall:{val:.2f}mm')

print('\n===CORRELATION ANALYSIS===')
correlation_matrix=np.corrcoef(rainfall,yield_data)
correlation=correlation_matrix[0,1]
print(f'Rainfall-Yield Correlation:{correlation:.6f}')
print('Interpretation:Strong positive correlation' if correlation>0.7 else 'Moderate correlation' if correlation>0.4 else 'Weak correlation')

print('\n===CUMULATIVE RAINFALL===')
cumulative=np.cumsum(rainfall)
print('Month\tRainfall\tCumulative ')
for month,(rain,cum) in enumerate(zip(rainfall,cumulative),1):
  print(f'{month}\t{rain}\t\t{cum}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_coords=(rainfall)
y_coords=(yield_data)
plt.plot(x_coords,y_coords)
plt.xlabel("rainfall(in mm)")
plt.ylabel("yield(in tons\ha)")
plt.show()

```

```python
import numpy as np
#fertilizer input for 5  years
n=5

fertilizer=[]
for i in range(n):
  fertilizer_input=int(input("enter fertilizer amount:"))
  fertilizer.append(fertilizer_input)
print(fertilizer)
yields=[]
for i in range(n):
  yield_data=float(input("enter yield:"))
  yields.append(yield_data)
print(yields)
```

```python
print('\n===YIELD STATISTICS===')
print(f'Mean:{np.mean(yields):.2f}tons/ha')
print(f'Median:{np.median(yields):.2f}tons/ha')
print(f'Std Dev:{np.std(yields):.2f}tons/ha')

print('\n===PERCENTILES===')
percentiles=[25,50,75,90]
fertilizer_percentiles=np.percentile(fertilizer,percentiles)
for p,val in zip(percentiles,fertilizer_percentiles):
  print(f'{p}th percentile fertilizer:{val:.2f}kg')

print('\n===CORRELATION ANALYSIS===')
correlation_matrix=np.corrcoef(fertilizer,yields)
correlation=correlation_matrix[0,1]
print(f'fertilizer-Yield Correlation:{correlation:.6f}')
print('Interpretation:Strong positive correlation' if correlation>0.7 else 'Moderate correlation' if correlation>0.4 else 'Weak correlation')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_coords=(fertilizer)
y_coords=(yields)
plt.plot(x_coords,y_coords)
plt.xlabel("fertilizer(in kg)")
plt.ylabel("yield(in tons\ha)")
plt.show()
```
