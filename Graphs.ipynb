```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.normal(170,10,250)
plt.hist(data,bins=40,color='red')
plt.xlabel("value")
plt.ylabel("frequency")
plt.title("Histogram example")
plt.show()
```

```python
data=[30,35,40,42,45,47,49,50,51,52,53,55,57,60,62,65,70,75]
plt.figure(figsize=(8,5))
plt.hist(
    data,bins=5,color="blue",edgecolor="black",alpha=1,density=True,range=(30,80)
)
plt.xlabel("value")
plt.ylabel("density")
plt.title("Histogram Without Random Data")
plt.grid(True)
plt.tight_layout()
plt.show()
```
