```python
def modify_inplace(nums):
  i=0
  while i<len(nums):
    if nums[i]<25:
      nums[i]*=2
    else:
      del nums[i]
      continue
    i+=1
  return nums
print(modify_inplace([10,20,30]))
```

```python
def nest_func(nested):
  for sublist in nested:
    if len(sublist)>1:
     print(sum(sublist))
nest_func([[1,2],[3]])
```

```python
def process_tuple(tup):
  total=0
  i=0
  while i<len(tup):
    if tup[i]%2==0:
      total+=tup[i]
    i+=1
  return total
process_tuple((10,20,30))
```

```python
scans=[[1,0,1],[0,1,0],[1,0,0]]
alert=[]
for i in range(len(scans)):
  row_count=0
  for j in range(len(scans[i])):
    if scans[i][j]==1:
      row_count+=1
  if row_count>1:
    alert.append(f"row{i}quarantine")
print(alert)
```
