```python
def load_data():
  """reads data and returns a list of dictionaries"""
  crops=[]
  file=open("crops.txt","r")
  for line in file:
    line=line.strip("\n")
    print(line)
    if line:
      data=line.split(",")
      crop={
          "crop_name":data[0],
          "season":data[1],
          "area":float(data[2]),
          "yield":float(data[3]),
          "cost":float(data[4]),
          "revenue":float(data[5])
    }
      crops.append(crop)

  return crops

seasonal_crops=load_data()
print(seasonal_crops)

def season(crops):
    """creates a list of kharif and rabi crops"""
    Kharif_list=[]
    Rabi_list=[]
    for crop in crops:
      if crop["season"]=="Kharif":
       Kharif_list.append(crop["crop_name"])
      else:
       Rabi_list.append(crop["crop_name"])
    print("Kharif Crops:",Kharif_list)
    print("Rabi Crops:",Rabi_list)

season(seasonal_crops)

def calculate_profit(crop):
  """calculates the profit"""
  return crop["revenue"]-crop["cost"]

def display_crops(crops):
  """displays profit of crops"""
  print("\nCrop Records:")
  for crop in crops:
    profit=calculate_profit(crop)
    print(crop["crop_name"],"-",round(profit,2))
  return profit
income=seasonal_crops
display_crops(income)

def calculate_average(crop):
  """returns average of crop yield"""
  return crop["yield"]/crop["area"]

def summary_statistics(crops):
  """displays the statistical summary of crops"""
  print("\nSTATISTICAL SUMMARY:")
  for crop in crops:
    average_yield=calculate_average(crop)
    print(crop["crop_name"],  "-" ,round(average_yield,2))
  return average_yield

statistics=income
summary_statistics(statistics)


```
