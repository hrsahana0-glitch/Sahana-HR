```python
squares={x:x*x for x in range(5)}
print(squares)
```

```python
ssquares={x:x*x for x in range(5)}
print(squares)
squares.update({5:25,6:36,7:49})
print("After update():",squares)
```

```python
last_item=squares.popitem()
print("removed item",last_item)
print("AFter pop item():",squares)
```

```python
removed_value=squares.pop(1,2)
print("removed value",removed_value)
print("after pop():",squares)
```

```python
squares={x:x*x for x in range(5)}
print(squares)
removed_value=squares.pop(0)
print("removed value",removed_value)
print("after pop():",squares)
```

```python
#nested dictionaries
student={
    "name":"MAHA",
    "marks":{"agronomy":49,"pathology":48.5}
}
print(student)
```
