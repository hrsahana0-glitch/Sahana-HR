```python
thislist=["apple","banana","cherry","blackcurrrant","banana"]

# To safely remove all occurrences of 'banana' using a for loop,
# we create a new list containing only the items that are not 'banana'.
new_list = []
for item in thislist:
    if item != "banana":
        new_list.append(item)

thislist = new_list
print(thislist)
```
