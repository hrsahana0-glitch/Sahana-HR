```python
import numpy as np

#Matrix Multiplication

A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
print(A)
print(B)
A@B
```

```python
#Element-wise Multiplication

result=A*B
print(result)
```

```python
#Transpose(swap rows and columns)

transposed=A.T
transposed=np.transpose(A)
print(transposed)
print(A)
```

```python
det=np.linalg.det(A)
print(det)
```

```python
#Matrix Inverse

inverse=np.linalg.inv(A)
print(inverse)
```

```python
#Matrix Rank

C=np.array([[1,2,3],[4,5,6]])
rank=np.linalg.matrix_rank(C)
print(rank)
print(C)
```

```python
eigenvectors=np.linalg.eig(A)
print(eigenvectors)
```

```python
#Solving linear systems

solution=np.linalg.solve(A,B)
print(solution)
```
