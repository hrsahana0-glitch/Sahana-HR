```python
#index() method
#Returns the index of the first occurence of an element
thislist=["apple","banana","cherry","banana"]
index_value=thislist.index ("cherry")
print("index of banana:",index_value)
```

```python
#count()method
#counts how many times an element appears in the list
thislist=["apple","banana","apple","cherry","apple"]
count_value=thislist.count("apple")
print("count of apple:",count_value)
```

```python
#reverse()method
#reverses the order of the elemnts in list(in place)
thislist=["orange","mango","kiwi","banana"]
print("original list:",thislist)
thislist.reverse()
print("reversed list:",thislist)
```

```python
original_list=[1,2,3,4,5]
reversed_iterator=reversed(original_list)
reversed_list=list(reversed_iterator)
print(reversed_list)
```

```python
#list length
#to determine how many items a list has,use the len() function:
thislist=["apple","banana","cherry"]
print(len(thislist))
```

```python
#list items_data types
#list items can be of any data type:
list1=["apple","banana","cherry"]
list2=[1,5,7,9,3]
list3=[True,False,False]
print(list1)
print(list2)
print(list3)
```

```python
#using len to access elements, but should be cautious while using it
for item in range(len(list1)):
  print(item)
```

```python
list1=["apple","banana","cherry"]
for item in range(len(list1)):
  print(list1[item])
```
